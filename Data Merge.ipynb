```python
import pandas as pd
import glob
import os
```

# Data Collection

```python
# folder_path = r"C:\Users\Ayu Sekar\Documents\GitHub\Jakarta-Weather-Prediction-2025\Dataset Raw"

folder_path = "Dataset Raw"



all_files = glob.glob(os.path.join(folder_path, "*.xlsx"))

df_list = []

for file in all_files:
    try:
        df = pd.read_excel(file)
        #Delete Header
        df = df.iloc[7:, :].reset_index(drop=True)
        #Reset index
        df.dropna(how='all', inplace=True)
        df.reset_index(drop=True, inplace=True)
        #Delete footer
        if len(df) > 13:
            df = df.iloc[:-13, :]

        #Reset index + delete 
        df.dropna(how='all', inplace=True)
        df.reset_index(drop=True, inplace=True)

        df_list.append(df)
        print(f"Berhasil proses: {os.path.basename(file)} | Baris tersisa: {len(df)}")

    except Exception as e:
        print(f"Gagal membaca {file}: {e}")

#Combine all excel
merged_df = pd.concat(df_list, ignore_index=True)

#Output
merged_df.to_excel("Datamergedoang.xlsx", index=False)
```
