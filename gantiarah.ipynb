````python
# =====================================================================
# COMPREHENSIVE DIAGNOSTIC: DATA QUALITY vs MODEL QUALITY
# Systematic framework to identify root cause of poor performance
# =====================================================================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
from sklearn.preprocessing import RobustScaler
from sklearn.metrics import mean_absolute_error, r2_score, mean_squared_error
from sklearn.model_selection import learning_curve, validation_curve
from xgboost import XGBRegressor
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.linear_model import LinearRegression
from sklearn.dummy import DummyRegressor
import warnings
warnings.filterwarnings('ignore')

print("="*80)
print("DIAGNOSTIC FRAMEWORK: DATA QUALITY vs MODEL QUALITY")
print("="*80)
print("\nSystematic investigation to determine root cause of poor performance\n")

# =====================================================================
# 1. LOAD & PREPARE DATA
# =====================================================================
df = pd.read_excel("updated_dataset_final - Copy.xlsx", parse_dates=['Tanggal'])
df = df.sort_values("Tanggal").reset_index(drop=True)

num_cols = [
    'Temperatur Minimum','Temperatur Maksimum','Temperatur Rata-rata',
    'Kelembapan Rata-rata','Curah Hujan (mm)',
    'Lamanya Penyinaran Matahari','Arah Angin Terbanyak (¬∞)'
]

for col in num_cols:
    df[col] = pd.to_numeric(df[col], errors="coerce")

df.replace(8888, np.nan, inplace=True)

# Add features
def add_features(df):
    df = df.copy()
    df["month"] = df["Tanggal"].dt.month
    df["month_sin"] = np.sin(2*np.pi*df["month"]/12)
    df["month_cos"] = np.cos(2*np.pi*df["month"]/12)
    df["temp_range"] = df["Temperatur Maksimum"] - df["Temperatur Minimum"]
    df["temp_hum"] = df["Temperatur Rata-rata"] * df["Kelembapan Rata-rata"]
    
    # Lags
    df["rain_lag1"] = df["Curah Hujan (mm)"].shift(1)
    df["rain_lag7"] = df["Curah Hujan (mm)"].shift(7)
    df["rain_mean_7d"] = df["Curah Hujan (mm)"].shift(1).rolling(7).mean()
    df["rain_mean_30d"] = df["Curah Hujan (mm)"].shift(1).rolling(30).mean()
    
    return df.dropna()

df_feat = add_features(df)

# Split
train = df_feat[df_feat['Tanggal'] < '2024-01-01']
val = df_feat[(df_feat['Tanggal'] >= '2024-01-01') & (df_feat['Tanggal'] < '2025-01-01')]
test = df_feat[df_feat['Tanggal'] >= '2025-01-01']

features = [
    'Temperatur Rata-rata', 'Kelembapan Rata-rata', 'temp_range', 'temp_hum',
    'rain_lag1', 'rain_lag7', 'rain_mean_7d', 'rain_mean_30d',
    'month_sin', 'month_cos'
]

X_train = train[features]
y_train = train['Curah Hujan (mm)']
X_val = val[features]
y_val = val['Curah Hujan (mm)']
X_test = test[features]
y_test = test['Curah Hujan (mm)']

scaler = RobustScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_val_scaled = scaler.transform(X_val)
X_test_scaled = scaler.transform(X_test)

# =====================================================================
# TEST 1: DATA QUALITY ASSESSMENT
# =====================================================================
print("\n" + "="*80)
print("TEST 1: DATA QUALITY ASSESSMENT")
print("="*80)

# 1.1 Missing values
print("\n[1.1] Missing Values Analysis:")
print("-" * 80)
missing = df[num_cols].isnull().sum()
missing_pct = (missing / len(df)) * 100
missing_df = pd.DataFrame({'Missing': missing, 'Percentage': missing_pct})
print(missing_df[missing_df['Missing'] > 0])

if missing.sum() > 0:
    print("‚ö†Ô∏è  WARNING: Missing values detected. This WILL hurt model performance.")
else:
    print("‚úì No missing values in features")

# 1.2 Target variable distribution
print("\n[1.2] Target Variable Distribution:")
print("-" * 80)
print(f"Mean: {y_train.mean():.2f} mm")
print(f"Median: {y_train.median():.2f} mm")
print(f"Std: {y_train.std():.2f} mm")
print(f"Skewness: {stats.skew(y_train):.2f}")
print(f"Kurtosis: {stats.kurtosis(y_train):.2f}")

zero_pct = (y_train == 0).mean() * 100
extreme_pct = (y_train > 50).mean() * 100
print(f"\nZero rainfall days: {zero_pct:.1f}%")
print(f"Extreme events (>50mm): {extreme_pct:.2f}%")

if stats.skew(y_train) > 2:
    print("‚ö†Ô∏è  SEVERE SKEWNESS: Target is highly skewed. Consider transformation or specialized models.")
if zero_pct > 50:
    print("‚ö†Ô∏è  SEVERE ZERO-INFLATION: >50% zeros. Standard regression will struggle.")
if extreme_pct < 1:
    print("‚ö†Ô∏è  EXTREME RARITY: <1% extreme events. Model will underpredict them.")

# 1.3 Feature-Target Correlations
print("\n[1.3] Feature-Target Correlations:")
print("-" * 80)
correlations = train[features + ['Curah Hujan (mm)']].corr()['Curah Hujan (mm)'].drop('Curah Hujan (mm)').sort_values(ascending=False)
print(correlations.head(10))

max_corr = correlations.abs().max()
print(f"\nStrongest correlation: {max_corr:.3f}")

if max_corr < 0.3:
    print("‚ö†Ô∏è  WEAK CORRELATIONS: Features have weak linear relationship with target.")
    print("   ‚Üí This suggests either:")
    print("      1. Missing important features")
    print("      2. Non-linear relationships (try non-linear models)")
    print("      3. Target is inherently unpredictable with these features")
else:
    print("‚úì Reasonable correlations exist")

# 1.4 Multicollinearity
print("\n[1.4] Multicollinearity Check (VIF):")
print("-" * 80)
from statsmodels.stats.outliers_influence import variance_inflation_factor

vif_data = pd.DataFrame()
vif_data["Feature"] = features
vif_data["VIF"] = [variance_inflation_factor(X_train.values, i) for i in range(len(features))]
vif_data = vif_data.sort_values('VIF', ascending=False)
print(vif_data.head())

high_vif = vif_data[vif_data['VIF'] > 10]
if len(high_vif) > 0:
    print(f"\n‚ö†Ô∏è  HIGH MULTICOLLINEARITY: {len(high_vif)} features with VIF > 10")
    print("   ‚Üí This can hurt model interpretability but usually OK for prediction")
else:
    print("‚úì No severe multicollinearity")

# =====================================================================
# TEST 2: INTRINSIC PREDICTABILITY (UPPER BOUND)
# =====================================================================
print("\n" + "="*80)
print("TEST 2: INTRINSIC PREDICTABILITY (Theoretical Upper Bound)")
print("="*80)

# 2.1 Nearest Neighbor baseline (simplest possible model)
print("\n[2.1] K-Nearest Neighbors Baseline:")
print("-" * 80)
from sklearn.neighbors import KNeighborsRegressor

knn = KNeighborsRegressor(n_neighbors=5)
knn.fit(X_train_scaled, y_train)
pred_knn = knn.predict(X_test_scaled)
pred_knn = np.maximum(pred_knn, 0)

r2_knn = r2_score(y_test, pred_knn)
mae_knn = mean_absolute_error(y_test, pred_knn)

print(f"KNN R¬≤: {r2_knn:.4f}")
print(f"KNN MAE: {mae_knn:.2f} mm")

# 2.2 Signal-to-Noise Ratio
print("\n[2.2] Signal-to-Noise Ratio:")
print("-" * 80)

# Estimate noise using residuals from best simple model
lr = LinearRegression()
lr.fit(X_train_scaled, y_train)
residuals = y_train - lr.predict(X_train_scaled)
noise_std = residuals.std()
signal_std = y_train.std()
snr = signal_std / noise_std

print(f"Signal Std: {signal_std:.2f}")
print(f"Noise Std: {noise_std:.2f}")
print(f"SNR: {snr:.2f}")

if snr < 2:
    print("‚ö†Ô∏è  LOW SIGNAL-TO-NOISE RATIO: Noise dominates signal.")
    print("   ‚Üí This is a FUNDAMENTAL DATA PROBLEM")
    print("   ‚Üí Even perfect models will have low R¬≤")
else:
    print("‚úì Reasonable signal-to-noise ratio")

# 2.3 Permutation Test (Is there ANY signal?)
print("\n[2.3] Permutation Test (Is the relationship real?):")
print("-" * 80)

# Train on real data
real_score = r2_knn

# Train on shuffled data (null hypothesis)
null_scores = []
for i in range(100):
    y_shuffled = y_train.sample(frac=1, random_state=i).values
    knn.fit(X_train_scaled, y_shuffled)
    pred = knn.predict(X_test_scaled)
    null_scores.append(r2_score(y_test, pred))

null_mean = np.mean(null_scores)
p_value = (np.sum(np.array(null_scores) >= real_score) + 1) / (len(null_scores) + 1)

print(f"Real R¬≤: {real_score:.4f}")
print(f"Null R¬≤ (shuffled): {null_mean:.4f}")
print(f"p-value: {p_value:.4f}")

if p_value < 0.05:
    print("‚úì Relationship is statistically significant (not random)")
else:
    print("‚ö†Ô∏è  WARNING: Relationship may be random!")

# =====================================================================
# TEST 3: MODEL QUALITY ASSESSMENT
# =====================================================================
print("\n" + "="*80)
print("TEST 3: MODEL QUALITY ASSESSMENT")
print("="*80)

# 3.1 Multiple models comparison
print("\n[3.1] Multiple Model Comparison:")
print("-" * 80)

models = {
    'Baseline (Mean)': DummyRegressor(strategy='mean'),
    'Linear Regression': LinearRegression(),
    'Random Forest': RandomForestRegressor(n_estimators=100, random_state=42, max_depth=10),
    'Gradient Boosting': GradientBoostingRegressor(n_estimators=100, random_state=42),
    'XGBoost': XGBRegressor(n_estimators=100, random_state=42)
}

results = []
for name, model in models.items():
    model.fit(X_train_scaled, y_train)
    pred = model.predict(X_test_scaled)
    pred = np.maximum(pred, 0)
    
    r2 = r2_score(y_test, pred)
    mae = mean_absolute_error(y_test, pred)
    rmse = np.sqrt(mean_squared_error(y_test, pred))
    
    results.append({
        'Model': name,
        'R¬≤': r2,
        'MAE': mae,
        'RMSE': rmse
    })

results_df = pd.DataFrame(results).sort_values('R¬≤', ascending=False)
print(results_df.to_string(index=False))

# Check if complex models are much better
best_r2 = results_df.iloc[0]['R¬≤']
baseline_r2 = results_df[results_df['Model'] == 'Baseline (Mean)']['R¬≤'].values[0]
improvement = best_r2 - baseline_r2

print(f"\nImprovement over baseline: {improvement:.4f}")

if improvement < 0.1:
    print("‚ö†Ô∏è  MINIMAL IMPROVEMENT: Complex models barely beat baseline.")
    print("   ‚Üí This is a DATA PROBLEM, not a model problem")
    print("   ‚Üí Features don't contain enough information")
elif best_r2 < 0.3:
    print("‚ö†Ô∏è  LOW ABSOLUTE PERFORMANCE: Even best model has low R¬≤")
    print("   ‚Üí This suggests FUNDAMENTAL LIMITS in the data")
else:
    print("‚úì Models show meaningful improvement")

# 3.2 Learning Curves
print("\n[3.2] Learning Curves (Bias-Variance Diagnosis):")
print("-" * 80)

xgb_model = XGBRegressor(n_estimators=200, learning_rate=0.05, max_depth=6, random_state=42)

train_sizes, train_scores, val_scores = learning_curve(
    xgb_model, X_train_scaled, y_train,
    train_sizes=np.linspace(0.1, 1.0, 10),
    cv=5,
    scoring='r2',
    n_jobs=-1
)

train_mean = train_scores.mean(axis=1)
val_mean = val_scores.mean(axis=1)

final_train_r2 = train_mean[-1]
final_val_r2 = val_mean[-1]
gap = final_train_r2 - final_val_r2

print(f"Training R¬≤ (full data): {final_train_r2:.4f}")
print(f"Validation R¬≤ (CV): {final_val_r2:.4f}")
print(f"Train-Val Gap: {gap:.4f}")

if gap > 0.2:
    print("‚ö†Ô∏è  HIGH VARIANCE: Model is overfitting")
    print("   ‚Üí This is a MODEL PROBLEM")
    print("   ‚Üí Try: regularization, simpler models, more data")
elif final_train_r2 < 0.5:
    print("‚ö†Ô∏è  HIGH BIAS: Model underfits even training data")
    print("   ‚Üí This is a DATA PROBLEM")
    print("   ‚Üí Features lack predictive power")
else:
    print("‚úì Reasonable bias-variance tradeoff")

# 3.3 Validation Curves (Hyperparameter sensitivity)
print("\n[3.3] Validation Curve (Hyperparameter Sensitivity):")
print("-" * 80)

param_range = [3, 5, 7, 9, 11]
train_scores, val_scores = validation_curve(
    XGBRegressor(n_estimators=100, random_state=42),
    X_train_scaled, y_train,
    param_name='max_depth',
    param_range=param_range,
    cv=5,
    scoring='r2'
)

val_mean = val_scores.mean(axis=1)
best_depth_idx = val_mean.argmax()
improvement_from_tuning = val_mean.max() - val_mean.min()

print(f"Best max_depth: {param_range[best_depth_idx]}")
print(f"R¬≤ range: {val_mean.min():.4f} to {val_mean.max():.4f}")
print(f"Improvement from tuning: {improvement_from_tuning:.4f}")

if improvement_from_tuning < 0.05:
    print("‚ö†Ô∏è  LOW SENSITIVITY: Hyperparameters don't matter much")
    print("   ‚Üí This is a DATA PROBLEM")
    print("   ‚Üí No amount of tuning will help significantly")
else:
    print("‚úì Hyperparameters have meaningful impact")

# =====================================================================
# TEST 4: DATA SUFFICIENCY
# =====================================================================
print("\n" + "="*80)
print("TEST 4: DATA SUFFICIENCY")
print("="*80)

# 4.1 Sample size analysis
print("\n[4.1] Sample Size Analysis:")
print("-" * 80)

n_features = len(features)
n_samples = len(X_train)
ratio = n_samples / n_features

print(f"Number of features: {n_features}")
print(f"Number of samples: {n_samples}")
print(f"Samples per feature: {ratio:.1f}")

if ratio < 10:
    print("‚ö†Ô∏è  TOO FEW SAMPLES: Ratio < 10")
    print("   ‚Üí High risk of overfitting")
elif ratio < 20:
    print("‚ö†Ô∏è  BORDERLINE: Ratio < 20")
    print("   ‚Üí Use regularization")
else:
    print("‚úì Sufficient samples per feature")

# 4.2 Dimensionality analysis
print("\n[4.2] Effective Dimensionality (PCA):")
print("-" * 80)
from sklearn.decomposition import PCA

pca = PCA()
pca.fit(X_train_scaled)
cumsum = np.cumsum(pca.explained_variance_ratio_)

n_components_90 = np.argmax(cumsum >= 0.90) + 1
n_components_95 = np.argmax(cumsum >= 0.95) + 1

print(f"Components for 90% variance: {n_components_90} / {n_features}")
print(f"Components for 95% variance: {n_components_95} / {n_features}")

if n_components_90 < n_features * 0.5:
    print("‚úì Features are redundant. Dimensionality reduction may help.")
else:
    print("‚Üí Most features are unique. They all carry information.")

# =====================================================================
# TEST 5: TEMPORAL STABILITY
# =====================================================================
print("\n" + "="*80)
print("TEST 5: TEMPORAL STABILITY (Non-Stationarity)")
print("="*80)

# 5.1 Performance over time
print("\n[5.1] Performance Degradation Over Time:")
print("-" * 80)

years = sorted(df_feat['Tanggal'].dt.year.unique())[-5:]
temporal_results = []

for test_year in years:
    train_mask = df_feat['Tanggal'].dt.year < test_year
    test_mask = df_feat['Tanggal'].dt.year == test_year
    
    if train_mask.sum() < 365 or test_mask.sum() < 30:
        continue
    
    X_tr = scaler.fit_transform(df_feat[train_mask][features])
    y_tr = df_feat[train_mask]['Curah Hujan (mm)']
    X_te = scaler.transform(df_feat[test_mask][features])
    y_te = df_feat[test_mask]['Curah Hujan (mm)']
    
    model = XGBRegressor(n_estimators=100, random_state=42)
    model.fit(X_tr, y_tr)
    pred = np.maximum(model.predict(X_te), 0)
    
    r2 = r2_score(y_te, pred)
    temporal_results.append({'Year': test_year, 'R¬≤': r2})

if len(temporal_results) > 2:
    temporal_df = pd.DataFrame(temporal_results)
    print(temporal_df.to_string(index=False))
    
    # Check for trend
    correlation = stats.spearmanr(temporal_df['Year'], temporal_df['R¬≤'])
    print(f"\nCorrelation (Year vs R¬≤): œÅ={correlation.correlation:.3f}, p={correlation.pvalue:.4f}")
    
    if correlation.pvalue < 0.1 and correlation.correlation < 0:
        print("‚ö†Ô∏è  PERFORMANCE DEGRADING OVER TIME")
        print("   ‚Üí This is a DATA PROBLEM (non-stationarity)")
        print("   ‚Üí The relationships are changing over time")
    else:
        print("‚úì Performance is stable over time")

# =====================================================================
# FINAL DIAGNOSIS
# =====================================================================
print("\n" + "="*80)
print("FINAL DIAGNOSIS")
print("="*80)

# Scoring system
data_issues = 0
model_issues = 0

# Data quality checks
if max_corr < 0.3: data_issues += 2
if zero_pct > 50: data_issues += 2
if extreme_pct < 1: data_issues += 1
if stats.skew(y_train) > 2: data_issues += 1
if snr < 2: data_issues += 3
if improvement < 0.1: data_issues += 3
if final_train_r2 < 0.5: data_issues += 2
if improvement_from_tuning < 0.05: data_issues += 2

# Model quality checks
if gap > 0.2: model_issues += 2
if ratio < 20: model_issues += 1

print("\nüìä DIAGNOSTIC SCORE:")
print("-" * 80)
print(f"Data Issues Score: {data_issues} / 16")
print(f"Model Issues Score: {model_issues} / 3")

print("\nüîç VERDICT:")
print("="*80)

if data_issues > 10:
    verdict = "PRIMARY PROBLEM: DATA QUALITY"
    explanation = """
The root cause is FUNDAMENTAL DATA LIMITATIONS, not model choice.

Evidence:
‚Ä¢ Weak feature-target correlations
‚Ä¢ Low signal-to-noise ratio  
‚Ä¢ Minimal improvement over baseline
‚Ä¢ Low training performance (high bias)
‚Ä¢ Insensitivity to hyperparameters

RECOMMENDATIONS:
1. ‚úÖ Collect better/more features (atmospheric pressure, radar, satellite)
2. ‚úÖ Consider this a feature engineering problem, not a modeling problem
3. ‚úÖ Try domain-specific approaches (physics-informed ML)
4. ‚úÖ Set realistic expectations with stakeholders
5. ‚úÖ Focus on uncertainty quantification over point predictions
6. ‚ùå Don't waste time on complex models or extensive tuning

PORTFOLIO MESSAGE:
"I discovered the predictability limit comes from data insufficiency,
not modeling approach. This insight saved weeks of fruitless tuning."
    """
elif model_issues > data_issues:
    verdict = "PRIMARY PROBLEM: MODEL QUALITY"
    explanation = """
The model is the bottleneck, not the data.

Evidence:
‚Ä¢ High train-validation gap (overfitting)
‚Ä¢ Good training performance but poor test performance
‚Ä¢ Insufficient samples per feature

RECOMMENDATIONS:
1. ‚úÖ Apply regularization (L1, L2)
2. ‚úÖ Reduce model complexity
3. ‚úÖ Use cross-validation properly
4. ‚úÖ Collect more training samples
5. ‚úÖ Try ensemble methods
6. ‚úÖ Feature selection to reduce dimensionality

PORTFOLIO MESSAGE:
"Identified overfitting as root cause and applied regularization
techniques to improve generalization."
    """
else:
    verdict = "MIXED: BOTH DATA AND MODEL ISSUES"
    explanation = """
Both data quality and model implementation need improvement.

RECOMMENDATIONS:
1. ‚úÖ Improve data collection (more features, better sensors)
2. ‚úÖ Apply regularization to reduce overfitting
3. ‚úÖ Try different model architectures
4. ‚úÖ Consider ensemble approaches
5. ‚úÖ Implement online learning for temporal drift

PORTFOLIO MESSAGE:
"Systematically diagnosed both data and model limitations,
and proposed comprehensive improvement strategy."
    """

print(f"\n‚ö° {verdict}")
print(explanation)

# =====================================================================
# VISUALIZATION
# =====================================================================
fig, axes = plt.subplots(3, 3, figsize=(18, 14))

# Plot 1: Target distribution
ax = axes[0, 0]
ax.hist(y_train[y_train > 0], bins=50, edgecolor='black', alpha=0.7)
ax.set_xlabel('Rainfall (mm)', fontweight='bold')
ax.set_ylabel('Frequency', fontweight='bold')
ax.set_title('Target Distribution (non-zero)', fontweight='bold')
ax.axvline(y_train.mean(), color='red', linestyle='--', linewidth=2, label=f'Mean: {y_train.mean():.1f}')
ax.legend()
ax.grid(True, alpha=0.3, axis='y')

# Plot 2: Feature correlations
ax = axes[0, 1]
corr_sorted = correlations.abs().sort_values(ascending=True)
ax.barh(range(len(corr_sorted)), corr_sorted.values)
ax.set_yticks(range(len(corr_sorted)))
ax.set_yticklabels(corr_sorted.index, fontsize=8)
ax.set_xlabel('|Correlation|', fontweight='bold')
ax.set_title('Feature-Target Correlations', fontweight='bold')
ax.axvline(0.3, color='red', linestyle='--', linewidth=2, alpha=0.5, label='Weak threshold')
ax.legend()
ax.grid(True, alpha=0.3, axis='x')

# Plot 3: Model comparison
ax = axes[0, 2]
ax.bar(range(len(results_df)), results_df['R¬≤'], edgecolor='black', linewidth=1.5)
ax.set_xticks(range(len(results_df)))
ax.set_xticklabels(results_df['Model'], rotation=45, ha='right', fontsize=9)
ax.set_ylabel('R¬≤ Score', fontweight='bold')
ax.set_title('Model Comparison', fontweight='bold')
ax.grid(True, alpha=0.3, axis='y')
for i, v in enumerate(results_df['R¬≤']):
    ax.text(i, v + 0.02, f'{v:.3f}', ha='center', fontweight='bold', fontsize=9)

# Plot 4: Learning curve
ax = axes[1, 0]
ax.plot(train_sizes, train_mean, 'o-', linewidth=2, label='Training', markersize=8)
ax.plot(train_sizes, val_mean, 's-', linewidth=2, label='Validation', markersize=8)
ax.fill_between(train_sizes, 
                train_scores.mean(axis=1) - train_scores.std(axis=1),
                train_scores.mean(axis=1) + train_scores.std(axis=1),
                alpha=0.2)
ax.fill_between(train_sizes,
                val_scores.mean(axis=1) - val_scores.std(axis=1),
                val_scores.mean(axis=1) + val_scores.std(axis=1),
                alpha=0.2)
ax.set_xlabel('Training Size', fontweight='bold')
ax.set_ylabel('R¬≤ Score', fontweight='bold')
ax.set_title('Learning Curve (Bias-Variance)', fontweight='bold')
ax.legend()
ax.grid(True, alpha=0.3)

# Plot 5: Validation curve
ax = axes[1, 1]
train_mean_val = train_scores.mean(axis=1)
val_mean_val = val_scores.mean(axis=1)
ax.plot(param_range, train_mean_val, 'o-', linewidth=2, label='Training', markersize=8)
ax.plot(param_range, val_mean_val, 's-', linewidth=2, label='Validation', markersize=8)
ax.set_xlabel('max_depth', fontweight='bold')
ax.set_ylabel('R¬≤ Score', fontweight='bold')
ax.set_title('Validation Curve (Hyperparameter)', fontweight='bold')
ax.legend()
ax.grid(True, alpha=0.3)

# Plot 6: Permutation test
ax = axes[1, 2]
ax.hist(null_scores, bins=30, alpha=0.7, edgecolor='black', label='Null (shuffled)')
ax.axvline(real_score, color='red', linestyle='--', linewidth=3, label=f'Real: {real_score:.3f}')
ax.set_xlabel('R¬≤ Score', fontweight='bold')
ax.set_ylabel('Frequency', fontweight='bold')
ax.set_title('Permutation Test', fontweight='bold')
ax.legend()
ax.grid(True, alpha=0.3, axis='y')

# Plot 7: Temporal stability
if len(temporal_results) > 2:
    ax = axes[2, 0]
    temporal_df = pd.DataFrame(temporal_results)
    ax.plot(temporal_df['Year'], temporal_df['R¬≤'], 'o-', linewidth=3, markersize=10, color='crimson')
    ax.set_xlabel('Year', fontweight='bold')
    ax.set_ylabel('R¬≤ Score', fontweight='bold')
    ax.set_title('Performance Over Time', fontweight='bold')
    ax.grid(True, alpha=0.3)
    
    # Trend line
    z = np.polyfit(temporal_df['Year'], temporal_df['R¬≤'], 1)
    p = np.poly1d(z)
    ax.plot(temporal_df['Year'], p(temporal_df['Year']), 'k--', linewidth=2, alpha=0.7)
else:
    axes[2, 0].text(0.5, 0.5, 'Insufficient temporal data', 
                    ha='center', va='center', transform=axes[2, 0].transAxes)
    axes[2, 0].axis('off')

# # Plot 8: PCA variance
# ax = axes[2, 1]
# ax.plot(range(1, len(cumsum)+1), cumsum, 'o-', linewidth=2, markersize=8)
# ax.axhline(0.9, color='red', linestyle='--', linewidth=2, alpha=0.5, label='90%')
# ax.axhline(0.95, color='orange', linestyle='--', linewidth=2, alpha=0.5, label='95%')
# ax.set_xlabel('Number of Components', fontweight='bold')
# ax.set_ylabel('Cum
````
